```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins['island'].value_counts()
```

```python
penguins['species'].value_counts()
```

```python
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'island')
```

```python
# The ways of styling themes are as follows:

#  white
#  dark
#  whitegrid
#  darkgrid
#  ticks

# set_context() allows us to override parameters.This effects things like the size of the labels, lines, and other elements of the plot, but not the overall style -> set_style(). The context are
# poster
# paper - Ideal for printing in academic papers.
# notebook - Good for working inside Jupyter notebooks (default setting).
# talk - Good for presentation slides where readability matters.

```

```python
sns.set_style('white')
sns.set_context('notebook')
sns.scatterplot(data = penguins, x = 'flipper_length_mm', y = 'body_mass_g', hue = 'island', palette = 'Set1')
sns.despine(left = True)
```

```python
sns.scatterplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'island', style = 'sex', alpha = 0.5)
```

```python
sns.stripplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'island', dodge = True, jitter = True)
```

```python
# Swarm Plot
sns.set_context('talk')
sns.swarmplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'island')
sns.despine()
```

```python
sns.set_context('paper')
sns.histplot(data = penguins, x = 'body_mass_g', hue = 'sex', multiple = 'stack')
plt.show()
```

```python
# regression plot
sns.regplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', color = 'green')
```

```python
# line plot
sns.set_style('whitegrid')
sns.lineplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', style = 'sex', hue = 'island')
```

```python
# joint plot
sns.jointplot(data = penguins, y = 'flipper_length_mm', x = 'body_mass_g')
```

```python
sns.jointplot(data = penguins, y = 'flipper_length_mm', x = 'body_mass_g',  kind = 'reg')
```

```python
# bar plot
sns.barplot(data = penguins, x = 'island', y = 'body_mass_g',hue = 'sex', palette = ['pink', 'skyblue'], estimator = np.sum)
```

```python
# count plot
sns.countplot(data = penguins, x = 'species',hue = 'island', palette = 'Set2')
```

```python
# box plot
sns.boxplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'sex', palette = 'Set1')
```

```python
# violin plot
sns.violinplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'sex', split = True)
```

```python
sns.violinplot(data = penguins, x = 'species', y = 'body_mass_g', palette = 'pastel')
sns.swarmplot(data = penguins, x = 'species', y = 'body_mass_g', hue = 'island', palette = 'dark',size = 3)
```

```python
# KDE -> for estimating the probability density function of continuous or non-parametric data
# Or in simple terms: a smooth curve that shows the distribution of data - like a smoothed out version of a histogram.
# It helps you see where the data is concentrated (peaks) and how it spreads.
```

```python
sns.kdeplot(data= penguins, x = 'body_mass_g', hue = 'species', fill = True)
```

```python
# Heatmap: a 2D color-coded matrix that shows relationships or patterns between two variables.
# It's commonly used to:
# - Show correlation between features
#  - Visualize confusion matrices -> will se in the later stages
#  - Highlight missing data -> sns.heatmap(df.isnull())
```

```python
columns = ['bill_length_mm', 'flipper_length_mm', 'bill_depth_mm', 'body_mass_g']
penguins[columns].corr()
```

```python
# heatmap
sns.heatmap(data = penguins[columns].corr(), annot = True, cmap = 'Blues',vmin= -0.2, linewidth = '2', linecolor='black' )
```

```python
# Rug plot
sns.rugplot(data = penguins, y = 'body_mass_g', hue = 'species', palette = 'Set1', height = 0.3)
```

```python
# pair plot
sns.pairplot(data = penguins, hue = 'sex', palette = 'Set2', diag_kind = 'hist')
```

```python
# pair grid -> subplot grid for plotting relationship in a dataset.
graph = sns.PairGrid(data = penguins, hue = 'sex', palette = 'Set2')
graph.map_upper(sns.scatterplot)
graph.map_lower(sns.kdeplot)
graph.map_diag(sns.histplot)
graph.add_legend()
```
